```python
#importing Dependencies
import numpy as np
from sklearn import linear_model,datasets,tree
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#preparing data
#Generationg radom points on x-axis
number_of_sample = 100
x = np.linspace(-np.pi,np.pi,number_of_sample)
y = 0.5*x + np.sin(x)+np.random.random(x.shape)
#print(x.shape[0])
#print("Input Feature",x)
#print("Output Feature",y)
plt.scatter(x,y,color="red")#plot y v/s x in dots
plt.xlabel("input feature")
plt.ylabel("output feature")
plt.show()
```

```python
#spliting the data into train,test,validation sets
random_indices = np.random.permutation(number_of_sample)
#print(random_indices)
#Training Set
x_train = x[random_indices[:70]]
y_train = y[random_indices[:70]]
#Validation Set
x_val = x[random_indices[70:85]]
y_val = y[random_indices[70:85]]
#Test set
x_test = x[random_indices[85:]]
y_test = y[random_indices[85:]]
```

```python
#fitting the data into the model
model = linear_model.LinearRegression()#Creating instance of linear model
#sklearn takes input as matrix so the we have to reshape the matrix into colloum matrix
x_train_for_fittng = np.matrix(x_train.reshape(len(x_train),1))
y_train_for_fittng = np.matrix(y_train.reshape(len(y_train),1))
#print(x_train_for_fittng)
#print(y_train_for_fittng)

#fit the data into the model
model.fit(x_train_for_fittng,y_train_for_fittng)
model_prediction = model.predict(x.reshape((len(x),1)))
#print(model_prediction)
plt.scatter(x_train,y_train,color="red")
plt.plot(x.reshape((len(x),1)),model_prediction,color="green")
```

```python
#Evaluating the model
mean_val_error = np.mean( (y_val - model.predict(x_val.reshape(len(x_val),1)))**2 )
mean_test_error = np.mean( (y_test - model.predict(x_test.reshape(len(x_test),1)))**2 )

print('Validation MSE: ', mean_val_error, '\nTest MSE: ', mean_test_error)
```
